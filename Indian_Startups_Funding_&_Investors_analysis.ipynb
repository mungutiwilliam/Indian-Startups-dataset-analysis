```python
# importing matplotlib and seaborn libraries and also setting up default sizes for the figures 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator
sns.set(rc = {'figure.figsize':(15,8)})
```

# 1. Which sector was invested on most by the top 5 investors ?

```python
# checking the top 5 sectors where the top 5 investors invested in 
sns.countplot(data = top_Investors[top_Investors['Investor'].isin(top_5_investors) & top_Investors['Sector'].isin(top_5_sectors)], x='Sector', order=['FinTech','EdTech','Information Technology & Services','SaaS startup','Food & Beverages'])
plt.title("Top 5 sectors where Top 5 ivestors invested on", fontsize=15)
plt.gcf().set_size_inches(20, 10)
```

From the dataset it can be seen that Fintech was the leading sector that was imvested on and financial Services was the least invested on. The plot shows the ranks from the highest to the lowest from the top 5 sectors

#2. which stage did most startups belong to 

```python
# plot to show the stages where the various startups invested on the by the top 5 investors belonged to  
sns.countplot(data = top_Investors [top_Investors['Sector'].isin(top_5_sectors) ], x = 'Stage', order=['Seed','Pre-series A','Series C','Series E','Series D','Series B','Pre-series A1','Pre-seed'])
plt.title('The Stages where startups invested on by top 5 inestors and belong to the top 5 sectors', fontsize=15)
plt.gcf().set_size_inches(20, 10)
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
```

Stage seed had the most number of start ups that were invested on by the top 5 investors and those that belong to the top 5 sectors 

#3. which which year did the top 5 investors invest most on start ups

```python
# startups that belonged to top 5 years startups were founded
sns.countplot(data = top_Investors[ top_Investors['Sector'].isin(top_5_sectors) ], x = 'Founded', order = ['2020','2019','2015','2017','2018'])
plt.title("Startups Founded in top 5 years startups were Founded", fontsize=15)
plt.gcf().set_size_inches(20, 10)
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```

From the plot it can be seen that the year 2020 was the leading year when most start ups were founded and the least to be founded in the top 5 was 2018

#4. which investor in the top 5 investors invested the most on the start ups in top 5 sectors

```python
sns.countplot(x ='Investor', data = top_Investors[(top_Investors['Investor'].isin(top_5_investors) & (top_Investors['Sector'].isin(top_5_sectors)))], order=['Inflection Point Ventures','Tiger Global','Venture Catalysts','Elevation Capital','Titan Capital'])
plt.title('Sectors the top5 Investors invested on', fontsize = 15)
plt.gcf().set_size_inches(20, 10)
plt.xticks(rotation=90)
```
